```
# 👉 Paste your full dataset here (RAW-036 to RAW-150)

data = [
    {"id":"RAW-036","prompt":"Explain the theory of relativity...","category":"Science & Technology","difficulty":"medium","response":"Einstein's theory...","response_length":1345,"quality_score":4.4,"split":"validation"},
    {"id":"RAW-037","prompt":"Write a function that implements merge sort...","category":"Programming","difficulty":"medium","response":"Merge sort is...","response_length":1678,"quality_score":4.5,"split":"validation"},
]

df = pd.DataFrame(data)
df.head()
```

```
# Simple validation
assert df.isnull().sum().sum() == 0, "❌ Ada data kosong!"
assert df["quality_score"].between(0,5).all(), "❌ Score tidak valid!"

print("✅ Data valid!")
```

```
# Save to CSV
file_name = "ai_dataset_validation.csv"
df.to_csv(file_name, index=False)

print(f"✅ Dataset saved as {file_name}")
```
